```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')

y = [10, 15, 15, 20, 10]
x = np.arange(1, 6)

bw = 0.25  #

# 
plt.bar(x, y, color='g', width=bw, label='product #1')

#  X  ()
y = [15, 25, 10, 15, 20]
plt.bar(x + bw, y, color='lightblue', width=bw, label='product #2')

#  X 
y = [10, 20, 15, 10, 15]
plt.bar(x + (2 * bw), y, hatch='..', color='lightgray', 
        width=bw, label='product #3')

# tick  X 
plt.xticks(x + bw, [2016, 2017, 2018, 2019, 2020])

plt.legend(loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt

font_options = {'family' : 'tahoma',
                'weight' : 'normal',
                'size' : 16}

plt.rc('font', **font_options)

data = [25, 20, 35, 10, 5]
lb = ['', '', '', '', '']
explode = [0, 0.1, 0, 0.2, 0] 

plt.pie(data, labels=lb, 
        explode=explode, startangle=30, 
        colors=['lightgray', '#ffffcc', 'pink', 'lightgreen', 'aqua'],
        shadow=True, autopct="%.1f%%")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig) 
#ax = plt.axes(projection='3d')

x = np.random.randint(10, 20, 5)
y = np.random.randint(30, 40, 5)
z = np.random.uniform(1, 5, 5)

ax.plot3D(x, y, z, 'ro--')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig) 
#ax = plt.axes(projection='3d')

x = np.random.randint(10, 20, 20)
y = np.random.randint(30, 40, 20)
z = np.random.uniform(1, 5, 20)

ax.scatter(x, y, z, color='green', marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
